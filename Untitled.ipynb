```python
#div = np.sqrt(ae['y_hat_sigma'].to_numpy().std()**2 + e['y_hat_sigma'].to_numpy().std()**2)
def plot_stuff(df):
    x = df['y_hat'].to_numpy()
    y = df['y_hat_sigma'].to_numpy()# / div

    #idx = np.argsort(x)

    # Define the bin width
    bin_width = 0.05

    # Create bins
    bins = np.arange(x.min(), x.max() + bin_width, bin_width)

    # Digitize the x values to find the bin they belong to
    bin_indices = np.digitize(x, bins)

    # Compute the average y values within each bin
    bin_means = [y[bin_indices == i].mean() for i in range(1, len(bins))]

    # Calculate the center of each bin for plotting
    bin_centers = 0.5 * (bins[1:] + bins[:-1])
    return bin_centers,bin_means

# Plotting
fig = plt.figure(figsize=(8,6))
plt.plot(*plot_stuff(e), marker='o',label='Epistemic Only',color='k')
plt.plot(*plot_stuff(ae),marker='o',label="Epistemic and Aleatoric",color='red')
plt.legend(fontsize=18)
plt.xlabel('Probability',fontsize=30,labelpad=14)
plt.ylabel(r'$\bar{\sigma}_{epi.}$',fontsize=30,labelpad=14)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)
plt.xlim(0,1)
plt.savefig('Comparison_Epistemic_func_prob.pdf',bbox_inches='tight')
plt.show()
```
